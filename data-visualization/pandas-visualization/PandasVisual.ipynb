````python
import pandas as pd
import numpy as np
%matplotlib inline
````

````python
df1 = pd.read_csv('./df1', index_col=0) # this is gonna make the first column become the index for your dataset.
df1.head()
````

````python
df2 = pd.read_csv('./df2')
df2.head()
````

````python
# you can create a plot using pandas as it uses matplotlib under the hood
# using the .hist() method you can plot a histogram of a data column
# and also you can pass normal matplotlib argumnets in the hist(args*)
df1['A'].hist(bins=30)
````

### Pandas has several plot types built into it and we are going to just bruefly go and show examples of how to create these plots.
All the plots can be done through a command off of the dataframe
1. call the dataframe itself.
2. call ```dataframe.plot(kind=type of the plot'hist', add the plot args*)```
2. another method is instead of specifying the kind of the plot you can call the plot kind directly >> ```datafram.hist(args of the histogram plot)```

````python
df1['A'].hist(bins=25) # calling hist() directly
````

````python
# another type of a plots is AreaPlot
df2.plot.area()
````

````python
# and you can specify the opacity (transparency)
df2.plot.area(alpha=0.4)
````

````python
# now lets create the BarPlot
df2.plot.bar()
# NOTE: the x axis here have the index values of the df2 dataframe
# So you maybe want to use this with a categorical index as this would make it easier to interpret the plot.
````

````python
# you can stack the values
df2.plot.bar(stacked=True)
````

### i think that it is better to use seaborn for this kind of plots unless you want to quickly collect general information about the dataset

````python
# Notice that Histogram are the ones that you are gonna use the most for quick analysis
df1['A'].plot.hist(bins=30)
````

````python
# Lets take a look at our df1 with the time series index
df1.head()
````

````python
# X: Columns to use for the horizontal axis.
# Either the location or the label of the columns to be used.
# By default, it will use the DataFrame indices.

# here also you can add the matplotlib arguments
df1.plot.line(y='B', figsize=(12,8), lw=1)
````

````python
# Lets check scatter plot
df1.plot.scatter(x='A', y='B', c='black', alpha=0.4)
````

````python
# with the scatter plot you can specify the color "c" argument to be 
# of another column:
df1.plot.scatter(x='A', y='B', c='C', cmap='Blues')
# and by this we are showing three levels of informations (A,B,C), C is shown by the intensity of the color. 3d plot
````

````python
# also you can pass 'C' as a value of the (s) argument <size> so the size reflects the value of the 'C' column
df1.plot.scatter(x='A', y='B',c='D' ,s=df1['C']*100)
````

````python
# you also can use the box plot with pandas as it is creating a box plot for each column
df2.plot.box()
````

````python
# and for bivarient (two variable data) you can do hexagonal pair plot
# 1- lets create a new dataframe for this to show up correctly
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.head()
````

````python
# 2- lets create the hexbin for our 'df' datafram we just created
df.plot.hexbin(x='a', y='b', figsize=(12,8))
````

````python
# the size of the hexagons can be enlarged by the argument (gridsize=25.......)
df.plot.hexbin(x='a', y='b', gridsize=25, figsize=(12,8), cmap='coolwarm')
````

````python
# Also you can create a kernel density estimation (KDE) plot
df['a'].plot.kde()
````

````python
# also you can create a KDE plot using <.density()> as follows
# and also you can create KDE for the entire DataFrame
df2.plot.density()
````
