# MATPLOTLIB:
Matplotlib is the "grandfather" library of data visualization with Python. It was created by John Hunter. He created it to try to replicate MatLab's (another programming language) plotting capabilities in Python. So if you happen to be familiar with matlab, matplotlib will feel natural to you.

It is an excellent 2D and 3D graphics library for generating scientific figures. 

Some of the major Pros of Matplotlib are:

* Generally easy to get started for simple plots
* Support for custom labels and texts
* Great control of every element in a figure
* High-quality output in many formats
* Very customizable in general

Matplotlib allows you to create reproducible figures programmatically. Let's learn how to use it! Before continuing this lecture, I encourage you just to explore the official Matplotlib web page: http://matplotlib.org/

## Different methods to create a plot using matplotlib
### 1- functional method

### 2- Objected Oriented Method to create matplotlib plots
1. create a figure object
2. we called add_axes() method on the figure and pass a list of axes [left, bottom, width, height]
3. plot on that set of axes

```python
# the main idea of OO method is to create figure object and call methods off of it
fig = plt.figure() # running this you will see that a figure object has been create
# adding axes to the figure >> axes = fig.add_axes([left, bottom, width, height])
axes = fig.add_axes([0.1,0.1,0.8,0.8]) # this is gonna give you empty plot
# now lets plot on these set of axes
axes.plot(x, y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title of Plot')
```

```python
# creating two sets of figures on the same canves
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
# 0.2 > 20 percent from left, 0.5 > 50 percent from the bottom,
# 0.4 > 40 percent of canves size width, 0.3 > 30 percent of canves height

axes1.plot(x,y, 'b')
axes1.set_title('Bigger Plot')

axes2.plot(y,x, 'r')
axes2.set_title('Smaller Plot')
```

```python
# creating a plank canvas
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.2,0.5,0.4,0.3])

# larger figure axes
ax1.plot(x,y, 'b*-')
ax1.set_xlabel('Big X label')
ax1.set_ylabel('Big Y label')
ax1.set_title('Larger Plot')

# smaller figure axes
ax2.plot(y,x, 'r--')
ax2.set_xlabel('Small X label')
ax2.set_ylabel('Small Y label')
ax2.set_title('Smaller Plot')
```

### Objected oriented method for subplots
## subplots()

The plt.subplots() object will act as a more automatic axis manager.

Basic use cases:

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots(nrows=2, ncols=3) #control the number of columns and the number of the rows in the plot
plt.tight_layout() # to get rid off the overlapping of the plots together
#(try with and without this line) and put this at the end of your plot statements always
```

A common issue with matplolib is overlapping subplots or figures. We ca use **fig.tight_layout()** or **plt.tight_layout()** method, which automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content:

```python
# now if lets find what axes in the cell above mean
axes # list of matplot axes that you created.
```

```python
# OOOh it is a list of several axes (iterable) that you can iterate over them
axes = axes.flatten() #If you want to iterate through the subplots in the way you do it now, by flattening axes first: 
for ax in axes:
    ax.plot(x,y)
fig # to display the figure object
```

```python
axes[1].plot(x,y, 'r*')
axes[1].set_xlabel('X label')
axes[1].set_ylabel('Y label')
axes[1].set_title('Title of subplot')

# now lets display the figure to see the changes (as you call the object in another cell)
fig 
```

### Figure size, aspect ratio,and DPI

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the `figsize` and `dpi` keyword arguments. 
* `figsize` is a tuple of the width and height of the figure in inches
* `dpi` is the dots-per-inch (pixel per inch). 

For example:

```python
# fig = plt.figure(figsize=(width, height of figure in inches), dpi=pixels_per_inch> 100)
fig = plt.figure(figsize=(5,2), dpi=100)
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, 'r--')
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
ax.set_title('Title of Plot')
```

```python
fig # calling the object from another cell
```

```python
# you can do the same thing with subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(3,2), dpi=100)
plt.tight_layout()
axes = axes.flatten()
for ax in axes:
    ax.plot(x, y)
```

### Legends
You can use the **label="label text"** keyword argument when plots or other objects are added to the figure, and then using the **legend** method without arguments to add the legend to the figure: 

```python
# here lets create a new figure and add legends to it
fig = plt.figure(figsize=(3,2), dpi=200)
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, 'r--', label='x squared')
ax.plot(x, x**3, 'b*-', label='x cubed')
ax.legend()
```

```python
# Lots of options....

ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner
ax.legend(loc=10) # center

# .. many more options are available
# you can also specify the exact loc=(left, bottom) inside a tuple

# Most common to choose
ax.legend(loc=0) # let matplotlib decide the optimal location
fig
```

## Setting colors, linewidths, linetypes

Matplotlib gives you *a lot* of options for customizing colors, linewidths, and linetypes. 

There is the basic MATLAB like syntax (which I would suggest you avoid using for more clairty sake:

### Colors with MatLab like syntax
With matplotlib, we can define the colors of lines and other graphical elements in a number of ways. First of all, we can use the MATLAB-like syntax where `'b'` means blue, `'g'` means green, etc. The MATLAB API for selecting line styles are also supported: where, for example, 'b.-' means a blue line with dots:

```python
fig = plt.figure(figsize=(3,2), dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, 'b.-') # blue lines of dots
ax.plot(x, x**3,'g--' ) # green lines of dashes
```

### Colors with the color= parameter
We can also define colors by their names or RGB hex codes and optionally provide an alpha value using the `color` and `alpha` keyword arguments. Alpha indicates opacity.

```python
ax.plot(x, x**2, color='blue', alpha=0.5) # half-transparant
ax.plot(x, x**3, color='#8B008B') # RGB hex code
fig
```

```python
# lets do the same with subplots
fig, ax = plt.subplots()
ax.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x+2, color="#8B008B")        # RGB hex code
ax.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```

### Line and marker styles
To change the line width, we can use the `linewidth` or `lw` keyword argument. The line style can be selected using the `linestyle` or `ls` keyword arguments:

```python
# line width 1 == default, 50 == 50 times the default.
ax.plot(x, y, linewidth=10, color='red', alpha=0.3)
# you don't need to type'linewidth' instead you can type only 'lw'
fig
```

```python
# Now lets see the different Line Styles in Matplotlib:
fig, ax = plt.subplots(figsize=(12,6))

# different line width
ax.plot(x, x+1, color='red', linewidth=0.25)
ax.plot(x, x+2, color='red', linewidth=0.50)
ax.plot(x, x+3, color='red', linewidth=1)
ax.plot(x, x+4, color='red', linewidth=2)

# possible lines types options '-' , '–' , '-.' , ':', 'steps'
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(x, x+12, color="blue", lw=1, ls='--', marker='1')

# you also can specify the Marker Size using the markersize argument as follow
# marker size and color
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

```python
# Now lets discuss the markers in detail (below there are alot of things that you can do with markers)
fig, ax = plt.subplots(figsize=(12,6))
ax.plot(x, y, lw=2, ls='-.', marker='o', markersize=20, markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='blue')
```

### Control over axis appearance
now we will look at controlling axis sizing properties in a matplotlib figure.
## Plot range
We can configure the ranges of the axes using the `set_ylim` and `set_xlim` methods in the axis object, or `axis('tight')` for automatically getting "tightly fitted" axes ranges:

```python
fig, ax = plt.subplots(ncols=3, nrows=1, figsize=(12,4))

ax[0].plot(x, x**2, x, x**3)
ax[0].set_title('Default ax ranges')

ax[1].plot(x, x**2, x, x**3)
ax[1].axis('tight')
ax[1].set_title('Tight axes')

ax[2].plot(x, x**2, x, x**3)
ax[2].set_ylim([0, 60]) # determine the limit of the y axis that are gonna be plotted
ax[2].set_xlim([2,5]) # determine the limit of the x axis that are gonna be plotted
ax[2].set_title('Custome axes range')
```
